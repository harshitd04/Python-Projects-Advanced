```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
clean_df.insert(1,'Spread', spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#Top 5 degrees with highest value in 90th percentile
clean_df.sort_values(by = 'Mid-Career 90th Percentile Salary',ascending = False)[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
#Degrees with Greatest Spreads
low_risk[['Undergraduate Major', 'Spread']].tail()
```

```python
clean_df.groupby('Group').count()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean(numeric_only=True)
```
